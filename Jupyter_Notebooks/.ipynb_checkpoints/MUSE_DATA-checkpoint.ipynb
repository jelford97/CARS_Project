```python
RApix=np.int(hdr['CRPIX1'])
DECpix=np.int(hdr['CRPIX2'])
RA=hdr['CRVAL1']
DEC=hdr['CRVAL2']
Xaxis=hdr['NAXIS1']
Yaxis=hdr['NAXIS2']
Cell=np.abs(hdr['CD1_1'])
```

```python
def dist(NX,NY,beamsize,RApixel,DECpixel,C):
    X=np.arange(NX)
    Y=np.arange(NY)
    RAaxis=np.abs(((X-RApixel)*C))
    DECaxis=np.abs(((Y-DECpixel)*C))
    im=np.empty([NY,NX])
    for i in range(0,NX):
        for j in range(0,NY):
            im[j,i]=np.sqrt(RAaxis[i]**2+DECaxis[j]**2)
    return im<beamsize
    
```

```python
imbeam=dist(Xaxis,Yaxis,0.2/3600,RApix,DECpix,Cell)
```

```python
Onuc=OIII*imbeam
```
